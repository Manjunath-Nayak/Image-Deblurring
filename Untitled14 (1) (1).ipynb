```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from skimage import metrics
import warnings
warnings.filterwarnings("ignore")

np.random.seed(42)

# ============================================================
# 1. CREATE SAME SYNTHETIC DATASET
# ============================================================
def create_synthetic_images(num_images=5, size=(256, 256)):
    images = []

    for i in range(num_images):
        img = np.zeros(size)

        if i == 0:
            # Checkerboard
            block_size = 32
            for r in range(0, size[0], block_size):
                for c in range(0, size[1], block_size):
                    if ((r // block_size) + (c // block_size)) % 2 == 0:
                        img[r:r+block_size, c:c+block_size] = 1.0

        elif i == 1:
            # Circular waves
            center_x, center_y = size[0] // 2, size[1] // 2
            y, x = np.ogrid[:size[0], :size[1]]
            dist = np.sqrt((x - center_y)**2 + (y - center_x)**2)
            img = (np.sin(dist / 10) + 1) / 2

        elif i == 2:
            # Rectangles
            img[50:200, 50:200] = 1.0
            img[100:150, 100:150] = 0.0

        elif i == 3:
            # Gradient
            img = np.linspace(0, 1, size[1])[np.newaxis, :].repeat(size[0], axis=0)

        else:
            # Random shapes
            for _ in range(10):
                r = np.random.randint(20, size[0]-20)
                c = np.random.randint(20, size[1]-20)
                radius = np.random.randint(10, 30)
                y, x = np.ogrid[:size[0], :size[1]]
                mask = (x - c)**2 + (y - r)**2 <= radius**2
                img[mask] = np.random.rand()

        images.append(img.astype(np.float64))

    return images


# ============================================================
# 2. BLUR + NOISE OPERATORS
# ============================================================
def create_gaussian_kernel(size=9, sigma=2.0):
    ax = np.arange(-size // 2 + 1., size // 2 + 1.)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-(xx**2 + yy**2) / (2. * sigma**2))
    return kernel / np.sum(kernel)

def blur_operator(image, kernel):
    return convolve2d(image, kernel, mode="same", boundary="symm")

def blur_adjoint(image, kernel):
    kernel_flipped = np.flip(kernel)
    return convolve2d(image, kernel_flipped, mode="same", boundary="symm")

def add_noise(image, noise_std=0.01):
    noise = np.random.randn(*image.shape) * noise_std
    return np.clip(image + noise, 0, 1)


# ============================================================
# 3. GRADIENT, DIVERGENCE, TV
# ============================================================
def gradient(image):
    gx = np.zeros_like(image)
    gy = np.zeros_like(image)

    gx[:, :-1] = image[:, 1:] - image[:, :-1]
    gy[:-1, :] = image[1:, :] - image[:-1, :]

    return gx, gy

def divergence(gx, gy):
    div_x = np.zeros_like(gx)
    div_y = np.zeros_like(gy)

    div_x[:, 0] = -gx[:, 0]
    div_x[:, 1:-1] = -(gx[:, 1:-1] - gx[:, :-2])
    div_x[:, -1] = gx[:, -2]

    div_y[0, :] = -gy[0, :]
    div_y[1:-1, :] = -(gy[1:-1, :] - gy[:-2, :])
    div_y[-1, :] = gy[-2, :]

    return div_x + div_y

def tv_norm(image):
    gx, gy = gradient(image)
    return np.sum(np.sqrt(gx**2 + gy**2))


# ============================================================
# 4. METRICS
# ============================================================
def compute_psnr(original, reconstructed):
    return metrics.peak_signal_noise_ratio(original, reconstructed, data_range=1.0)

def compute_ssim(original, reconstructed):
    return metrics.structural_similarity(original, reconstructed, data_range=1.0)


# ============================================================
# 5. L2 + GRADIENT DESCENT
# ============================================================
def l2_gd(blurred_noisy, kernel, lambda_l2=0.01, lr=0.1, max_iter=200):
    x = blurred_noisy.copy()
    history = []

    for k in range(max_iter):
        residual = blur_operator(x, kernel) - blurred_noisy
        grad = blur_adjoint(residual, kernel) + lambda_l2 * x

        obj = 0.5 * np.sum(residual**2) + 0.5 * lambda_l2 * np.sum(x**2)
        history.append(obj)

        x = x - lr * grad
        x = np.clip(x, 0, 1)

    return x, history


# ============================================================
# 6. L2 + FISTA
# ============================================================
def l2_fista(blurred_noisy, kernel, lambda_l2=0.01, L=1.0, max_iter=200):
    x = blurred_noisy.copy()
    y = x.copy()
    t = 1.0
    history = []

    for k in range(max_iter):
        residual = blur_operator(x, kernel) - blurred_noisy
        obj = 0.5 * np.sum(residual**2) + 0.5 * lambda_l2 * np.sum(x**2)
        history.append(obj)

        residual_y = blur_operator(y, kernel) - blurred_noisy
        grad = blur_adjoint(residual_y, kernel) + lambda_l2 * y

        x_new = y - (1.0 / L) * grad
        x_new = np.clip(x_new, 0, 1)

        t_new = (1.0 + np.sqrt(1.0 + 4.0 * t**2)) / 2.0
        y = x_new + ((t - 1.0) / t_new) * (x_new - x)

        x = x_new
        t = t_new

    return x, history


# ============================================================
# 7. TV + ADMM
# ============================================================
def tv_admm(blurred_noisy, kernel, lambda_tv=0.05, rho=1.0, max_iter=200):
    x = blurred_noisy.copy()

    z_x = np.zeros_like(x)
    z_y = np.zeros_like(x)
    u_x = np.zeros_like(x)
    u_y = np.zeros_like(x)

    history = []

    for k in range(max_iter):
        residual = blur_operator(x, kernel) - blurred_noisy
        obj = 0.5 * np.sum(residual**2) + lambda_tv * tv_norm(x)
        history.append(obj)

        rhs = blur_adjoint(blurred_noisy, kernel)
        rhs += rho * divergence(z_x - u_x, z_y - u_y)

        # approximate x-update using inner gradient steps
        for _ in range(10):
            Ax = blur_operator(x, kernel)
            ATAx = blur_adjoint(Ax, kernel)

            gx, gy = gradient(x)
            DTDx = divergence(gx, gy)

            lhs = ATAx + rho * DTDx
            res = lhs - rhs

            step = 1.0 / (1.0 + 4.0 * rho)
            x = x - step * res
            x = np.clip(x, 0, 1)

        gx, gy = gradient(x)
        v_x = gx + u_x
        v_y = gy + u_y

        v_mag = np.sqrt(v_x**2 + v_y**2)
        shrink = np.maximum(v_mag - lambda_tv / rho, 0.0) / (v_mag + 1e-10)

        z_x = shrink * v_x
        z_y = shrink * v_y

        gx, gy = gradient(x)
        u_x += gx - z_x
        u_y += gy - z_y

    return x, history


# ============================================================
# 8. TV + FISTA
# ============================================================
def prox_tv_chambolle(x, lambda_prox, num_iter=10):
    p_x = np.zeros_like(x)
    p_y = np.zeros_like(x)
    tau = 0.25

    for _ in range(num_iter):
        div_p = divergence(p_x, p_y)
        gx, gy = gradient(x / lambda_prox - div_p)

        p_x_new = p_x + tau * gx
        p_y_new = p_y + tau * gy

        p_mag = np.maximum(1.0, np.sqrt(p_x_new**2 + p_y_new**2))

        p_x = p_x_new / p_mag
        p_y = p_y_new / p_mag

    return x - lambda_prox * divergence(p_x, p_y)

def tv_fista(blurred_noisy, kernel, lambda_tv=0.05, L=1.0, max_iter=200):
    x = blurred_noisy.copy()
    y = x.copy()
    t = 1.0
    history = []

    for k in range(max_iter):
        residual = blur_operator(x, kernel) - blurred_noisy
        obj = 0.5 * np.sum(residual**2) + lambda_tv * tv_norm(x)
        history.append(obj)

        residual_y = blur_operator(y, kernel) - blurred_noisy
        grad = blur_adjoint(residual_y, kernel)

        z = y - (1.0 / L) * grad
        x_new = prox_tv_chambolle(z, lambda_tv / L, num_iter=10)
        x_new = np.clip(x_new, 0, 1)

        t_new = (1.0 + np.sqrt(1.0 + 4.0 * t**2)) / 2.0
        y = x_new + ((t - 1.0) / t_new) * (x_new - x)

        x = x_new
        t = t_new

    return x, history


# ============================================================
# 9. NEW EXTENSION: TV + PRIMAL-DUAL / CHAMBOLLE-POCK
# ============================================================
def tv_primal_dual(
    blurred_noisy,
    kernel,
    lambda_tv=0.05,
    tau=0.008,
    sigma=0.008,
    theta=1.0,
    max_iter=200
):
    x = blurred_noisy.copy()
    x_bar = x.copy()

    p_x = np.zeros_like(x)
    p_y = np.zeros_like(x)

    history = []

    for k in range(max_iter):
        residual = blur_operator(x, kernel) - blurred_noisy
        obj = 0.5 * np.sum(residual**2) + lambda_tv * tv_norm(x)
        history.append(obj)

        x_old = x.copy()

        # Dual update
        gx, gy = gradient(x_bar)

        p_x = p_x + sigma * gx
        p_y = p_y + sigma * gy

        # Projection onto TV dual ball
        norm_p = np.maximum(1.0, np.sqrt(p_x**2 + p_y**2) / lambda_tv)

        p_x = p_x / norm_p
        p_y = p_y / norm_p

        # Primal update
        data_grad = blur_adjoint(
            blur_operator(x, kernel) - blurred_noisy,
            kernel
        )

        div_p = divergence(p_x, p_y)

        x = x - tau * (data_grad + div_p)
        x = np.clip(x, 0, 1)

        # Extrapolation
        x_bar = x + theta * (x - x_old)

    return x, history


# ============================================================
# 10. RUN FULL COMPARISON
# ============================================================
kernel_size = 9
sigma = 2.0
noise_std = 0.01
lambda_l2 = 0.01
lambda_tv = 0.05
max_iter = 200

kernel = create_gaussian_kernel(kernel_size, sigma)
images = create_synthetic_images(num_images=5, size=(256, 256))

image_names = [
    "Checkerboard",
    "Circular Waves",
    "Rectangles",
    "Gradient",
    "Random Shapes"
]

methods = [
    "L2_GD",
    "L2_FISTA",
    "TV_ADMM",
    "TV_FISTA",
    "TV_PrimalDual"
]

all_results = {m: {"psnr": [], "ssim": [], "history": []} for m in methods}
visual_results = []

print("=" * 80)
print("CONVEX IMAGE DEBLURRING COMPARISON WITH TV PRIMAL-DUAL EXTENSION")
print("=" * 80)
print(f"Dataset: Synthetic images")
print(f"Images used: {image_names}")
print(f"Kernel: Gaussian {kernel_size}x{kernel_size}, sigma={sigma}")
print(f"Noise std: {noise_std}")
print(f"Iterations: {max_iter}")
print("=" * 80)

for idx, original in enumerate(images):
    print(f"\nProcessing Image {idx+1}/5: {image_names[idx]}")

    blurred = blur_operator(original, kernel)
    blurred_noisy = add_noise(blurred, noise_std)

    # L2 + GD
    rec, hist = l2_gd(blurred_noisy, kernel, lambda_l2, lr=0.1, max_iter=max_iter)
    all_results["L2_GD"]["psnr"].append(compute_psnr(original, rec))
    all_results["L2_GD"]["ssim"].append(compute_ssim(original, rec))
    all_results["L2_GD"]["history"].append(hist)
    print(f"  L2 + GD            PSNR={all_results['L2_GD']['psnr'][-1]:.2f}, SSIM={all_results['L2_GD']['ssim'][-1]:.4f}")

    # L2 + FISTA
    rec, hist = l2_fista(blurred_noisy, kernel, lambda_l2, L=1.0, max_iter=max_iter)
    all_results["L2_FISTA"]["psnr"].append(compute_psnr(original, rec))
    all_results["L2_FISTA"]["ssim"].append(compute_ssim(original, rec))
    all_results["L2_FISTA"]["history"].append(hist)
    print(f"  L2 + FISTA         PSNR={all_results['L2_FISTA']['psnr'][-1]:.2f}, SSIM={all_results['L2_FISTA']['ssim'][-1]:.4f}")

    # TV + ADMM
    rec_admm, hist_admm = tv_admm(blurred_noisy, kernel, lambda_tv, rho=1.0, max_iter=max_iter)
    all_results["TV_ADMM"]["psnr"].append(compute_psnr(original, rec_admm))
    all_results["TV_ADMM"]["ssim"].append(compute_ssim(original, rec_admm))
    all_results["TV_ADMM"]["history"].append(hist_admm)
    print(f"  TV + ADMM          PSNR={all_results['TV_ADMM']['psnr'][-1]:.2f}, SSIM={all_results['TV_ADMM']['ssim'][-1]:.4f}")

    # TV + FISTA
    rec_fista, hist_fista = tv_fista(blurred_noisy, kernel, lambda_tv, L=1.0, max_iter=max_iter)
    all_results["TV_FISTA"]["psnr"].append(compute_psnr(original, rec_fista))
    all_results["TV_FISTA"]["ssim"].append(compute_ssim(original, rec_fista))
    all_results["TV_FISTA"]["history"].append(hist_fista)
    print(f"  TV + FISTA         PSNR={all_results['TV_FISTA']['psnr'][-1]:.2f}, SSIM={all_results['TV_FISTA']['ssim'][-1]:.4f}")

    # TV + Primal-Dual
    rec_pd, hist_pd = tv_primal_dual(
        blurred_noisy,
        kernel,
        lambda_tv=lambda_tv,
        tau=0.02,
        sigma=0.02,
        theta=1.0,
        max_iter=max_iter
    )
    all_results["TV_PrimalDual"]["psnr"].append(compute_psnr(original, rec_pd))
    all_results["TV_PrimalDual"]["ssim"].append(compute_ssim(original, rec_pd))
    all_results["TV_PrimalDual"]["history"].append(hist_pd)
    print(f"  TV + Primal-Dual   PSNR={all_results['TV_PrimalDual']['psnr'][-1]:.2f}, SSIM={all_results['TV_PrimalDual']['ssim'][-1]:.4f}")

    if idx == 0:
        visual_results = [
            original,
            blurred_noisy,
            rec_admm,
            rec_fista,
            rec_pd
        ]


# ============================================================
# 11. SUMMARY TABLE
# ============================================================
print("\n" + "=" * 80)
print("SUMMARY STATISTICS ACROSS ALL 5 SYNTHETIC IMAGES")
print("=" * 80)
print(f"{'Method':<18} {'Avg PSNR':>12} {'Std PSNR':>12} {'Avg SSIM':>12} {'Std SSIM':>12}")
print("-" * 80)

for method in methods:
    avg_psnr = np.mean(all_results[method]["psnr"])
    std_psnr = np.std(all_results[method]["psnr"])
    avg_ssim = np.mean(all_results[method]["ssim"])
    std_ssim = np.std(all_results[method]["ssim"])

    print(f"{method:<18} {avg_psnr:>12.4f} {std_psnr:>12.4f} {avg_ssim:>12.4f} {std_ssim:>12.4f}")

best_psnr_method = max(methods, key=lambda m: np.mean(all_results[m]["psnr"]))
best_ssim_method = max(methods, key=lambda m: np.mean(all_results[m]["ssim"]))

print("\nKEY FINDINGS:")
print(f"Best PSNR method: {best_psnr_method} = {np.mean(all_results[best_psnr_method]['psnr']):.4f} dB")
print(f"Best SSIM method: {best_ssim_method} = {np.mean(all_results[best_ssim_method]['ssim']):.4f}")

print("\nTV solver comparison:")
print(f"TV_ADMM Avg PSNR        = {np.mean(all_results['TV_ADMM']['psnr']):.4f}")
print(f"TV_FISTA Avg PSNR       = {np.mean(all_results['TV_FISTA']['psnr']):.4f}")
print(f"TV_PrimalDual Avg PSNR  = {np.mean(all_results['TV_PrimalDual']['psnr']):.4f}")


# ============================================================
# 12. VISUALIZE FIRST IMAGE RESULT
# ============================================================
titles = [
    "Original",
    "Blurred + Noise",
    "TV + ADMM",
    "TV + FISTA",
    "TV + Primal-Dual"
]

plt.figure(figsize=(18, 4))
for i, img in enumerate(visual_results):
    plt.subplot(1, 5, i + 1)
    plt.imshow(img, cmap="gray", vmin=0, vmax=1)
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()


# ============================================================
# 13. CONVERGENCE PLOT FOR FIRST IMAGE
# ============================================================
plt.figure(figsize=(9, 6))

for method in methods:
    plt.semilogy(all_results[method]["history"][0], label=method)

plt.xlabel("Iteration")
plt.ylabel("Objective Value (log scale)")
plt.title("Convergence Comparison on First Synthetic Image")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from skimage import metrics
import warnings
warnings.filterwarnings("ignore")

np.random.seed(42)

# =========================
# Dataset
# =========================
def create_synthetic_images(num_images=5, size=(256, 256)):
    images = []

    for i in range(num_images):
        img = np.zeros(size)

        if i == 0:
            block_size = 32
            for r in range(0, size[0], block_size):
                for c in range(0, size[1], block_size):
                    if ((r // block_size) + (c // block_size)) % 2 == 0:
                        img[r:r+block_size, c:c+block_size] = 1.0

        elif i == 1:
            cx, cy = size[0] // 2, size[1] // 2
            y, x = np.ogrid[:size[0], :size[1]]
            dist = np.sqrt((x - cy)**2 + (y - cx)**2)
            img = (np.sin(dist / 10) + 1) / 2

        elif i == 2:
            img[50:200, 50:200] = 1.0
            img[100:150, 100:150] = 0.0

        elif i == 3:
            img = np.linspace(0, 1, size[1])[np.newaxis, :].repeat(size[0], axis=0)

        else:
            for _ in range(10):
                r = np.random.randint(20, size[0]-20)
                c = np.random.randint(20, size[1]-20)
                radius = np.random.randint(10, 30)
                y, x = np.ogrid[:size[0], :size[1]]
                mask = (x - c)**2 + (y - r)**2 <= radius**2
                img[mask] = np.random.rand()

        images.append(img.astype(np.float64))

    return images


# =========================
# Operators
# =========================
def create_gaussian_kernel(size=9, sigma=2.0):
    ax = np.arange(-size // 2 + 1., size // 2 + 1.)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-(xx**2 + yy**2) / (2. * sigma**2))
    return kernel / np.sum(kernel)

def blur_operator(image, kernel):
    return convolve2d(image, kernel, mode="same", boundary="symm")

def blur_adjoint(image, kernel):
    return convolve2d(image, np.flip(kernel), mode="same", boundary="symm")

def add_noise(image, noise_std=0.01):
    noise = np.random.randn(*image.shape) * noise_std
    return np.clip(image + noise, 0, 1)

def gradient(image):
    gx = np.zeros_like(image)
    gy = np.zeros_like(image)
    gx[:, :-1] = image[:, 1:] - image[:, :-1]
    gy[:-1, :] = image[1:, :] - image[:-1, :]
    return gx, gy

def divergence(gx, gy):
    div_x = np.zeros_like(gx)
    div_y = np.zeros_like(gy)

    div_x[:, 0] = -gx[:, 0]
    div_x[:, 1:-1] = -(gx[:, 1:-1] - gx[:, :-2])
    div_x[:, -1] = gx[:, -2]

    div_y[0, :] = -gy[0, :]
    div_y[1:-1, :] = -(gy[1:-1, :] - gy[:-2, :])
    div_y[-1, :] = gy[-2, :]

    return div_x + div_y

def tv_norm(image):
    gx, gy = gradient(image)
    return np.sum(np.sqrt(gx**2 + gy**2))

def compute_psnr(original, restored):
    return metrics.peak_signal_noise_ratio(original, restored, data_range=1.0)

def compute_ssim(original, restored):
    return metrics.structural_similarity(original, restored, data_range=1.0)


# =========================
# TV + FISTA
# =========================
def prox_tv_chambolle(x, lambda_prox, num_iter=10):
    p_x = np.zeros_like(x)
    p_y = np.zeros_like(x)
    tau = 0.25

    for _ in range(num_iter):
        div_p = divergence(p_x, p_y)
        gx, gy = gradient(x / lambda_prox - div_p)

        p_x_new = p_x + tau * gx
        p_y_new = p_y + tau * gy

        p_mag = np.maximum(1.0, np.sqrt(p_x_new**2 + p_y_new**2))

        p_x = p_x_new / p_mag
        p_y = p_y_new / p_mag

    return x - lambda_prox * divergence(p_x, p_y)

def tv_fista(blurred_noisy, kernel, lambda_tv=0.05, L=1.0, max_iter=200):
    x = blurred_noisy.copy()
    y = x.copy()
    t = 1.0
    history = []

    for k in range(max_iter):
        residual = blur_operator(x, kernel) - blurred_noisy
        obj = 0.5 * np.sum(residual**2) + lambda_tv * tv_norm(x)
        history.append(obj)

        residual_y = blur_operator(y, kernel) - blurred_noisy
        grad = blur_adjoint(residual_y, kernel)

        z = y - (1.0 / L) * grad
        x_new = prox_tv_chambolle(z, lambda_tv / L, num_iter=10)
        x_new = np.clip(x_new, 0, 1)

        t_new = (1.0 + np.sqrt(1.0 + 4.0 * t**2)) / 2.0
        y = x_new + ((t - 1.0) / t_new) * (x_new - x)

        x = x_new
        t = t_new

    return x, history


# =========================
# NEW METHOD: TV + Split Bregman
# =========================
def shrink_isotropic(gx, gy, threshold):
    mag = np.sqrt(gx**2 + gy**2)
    shrink = np.maximum(mag - threshold, 0.0) / (mag + 1e-10)
    return shrink * gx, shrink * gy

def tv_split_bregman(
    blurred_noisy,
    kernel,
    lambda_tv=0.05,
    mu=1.0,
    max_iter=200,
    inner_iter=10
):
    u = blurred_noisy.copy()

    dx = np.zeros_like(u)
    dy = np.zeros_like(u)

    bx = np.zeros_like(u)
    by = np.zeros_like(u)

    history = []

    for k in range(max_iter):
        residual = blur_operator(u, kernel) - blurred_noisy
        obj = 0.5 * np.sum(residual**2) + lambda_tv * tv_norm(u)
        history.append(obj)

        # u-update: approximate solution of quadratic subproblem
        rhs = blur_adjoint(blurred_noisy, kernel) + mu * divergence(dx - bx, dy - by)

        for _ in range(inner_iter):
            Au = blur_operator(u, kernel)
            ATAu = blur_adjoint(Au, kernel)

            ux, uy = gradient(u)
            DTDu = divergence(ux, uy)

            lhs = ATAu + mu * DTDu
            res = lhs - rhs

            step = 1.0 / (1.0 + 4.0 * mu)
            u = u - step * res
            u = np.clip(u, 0, 1)

        # d-update: shrinkage
        ux, uy = gradient(u)

        dx, dy = shrink_isotropic(
            ux + bx,
            uy + by,
            lambda_tv / mu
        )

        # Bregman update
        bx = bx + ux - dx
        by = by + uy - dy

    return u, history


# =========================
# Run comparison
# =========================
kernel_size = 9
sigma = 2.0
noise_std = 0.01
lambda_tv = 0.05
max_iter = 200

kernel = create_gaussian_kernel(kernel_size, sigma)
images = create_synthetic_images(num_images=5, size=(256, 256))

image_names = [
    "Checkerboard",
    "Circular Waves",
    "Rectangles",
    "Gradient",
    "Random Shapes"
]

methods = ["TV_FISTA", "TV_SplitBregman"]
results = {m: {"psnr": [], "ssim": [], "history": []} for m in methods}

visual_results = []

print("=" * 80)
print("TV-FISTA vs TV-SPLIT BREGMAN COMPARISON")
print("=" * 80)
print("Dataset: Same 5 synthetic images")
print("Images:", image_names)
print(f"Kernel: Gaussian {kernel_size}x{kernel_size}, sigma={sigma}")
print(f"Noise std: {noise_std}")
print(f"Iterations: {max_iter}")
print("=" * 80)

for idx, original in enumerate(images):
    print(f"\nProcessing Image {idx+1}/5: {image_names[idx]}")

    blurred = blur_operator(original, kernel)
    blurred_noisy = add_noise(blurred, noise_std)

    # TV + FISTA
    rec_fista, hist_fista = tv_fista(
        blurred_noisy,
        kernel,
        lambda_tv=lambda_tv,
        L=1.0,
        max_iter=max_iter
    )

    psnr_fista = compute_psnr(original, rec_fista)
    ssim_fista = compute_ssim(original, rec_fista)

    results["TV_FISTA"]["psnr"].append(psnr_fista)
    results["TV_FISTA"]["ssim"].append(ssim_fista)
    results["TV_FISTA"]["history"].append(hist_fista)

    print(f"  TV + FISTA          PSNR={psnr_fista:.2f}, SSIM={ssim_fista:.4f}")

    # TV + Split Bregman
    rec_sb, hist_sb = tv_split_bregman(
        blurred_noisy,
        kernel,
        lambda_tv=lambda_tv,
        mu=1.0,
        max_iter=max_iter,
        inner_iter=10
    )

    psnr_sb = compute_psnr(original, rec_sb)
    ssim_sb = compute_ssim(original, rec_sb)

    results["TV_SplitBregman"]["psnr"].append(psnr_sb)
    results["TV_SplitBregman"]["ssim"].append(ssim_sb)
    results["TV_SplitBregman"]["history"].append(hist_sb)

    print(f"  TV + Split Bregman  PSNR={psnr_sb:.2f}, SSIM={ssim_sb:.4f}")

    if idx == 0:
        visual_results = [
            original,
            blurred_noisy,
            rec_fista,
            rec_sb
        ]


# =========================
# Summary
# =========================
print("\n" + "=" * 80)
print("SUMMARY STATISTICS")
print("=" * 80)
print(f"{'Method':<20} {'Avg PSNR':>12} {'Std PSNR':>12} {'Avg SSIM':>12} {'Std SSIM':>12}")
print("-" * 80)

for method in methods:
    avg_psnr = np.mean(results[method]["psnr"])
    std_psnr = np.std(results[method]["psnr"])
    avg_ssim = np.mean(results[method]["ssim"])
    std_ssim = np.std(results[method]["ssim"])

    print(f"{method:<20} {avg_psnr:>12.4f} {std_psnr:>12.4f} {avg_ssim:>12.4f} {std_ssim:>12.4f}")

best_psnr = max(methods, key=lambda m: np.mean(results[m]["psnr"]))
best_ssim = max(methods, key=lambda m: np.mean(results[m]["ssim"]))

print("\nKEY FINDINGS:")
print(f"Best PSNR: {best_psnr} = {np.mean(results[best_psnr]['psnr']):.4f} dB")
print(f"Best SSIM: {best_ssim} = {np.mean(results[best_ssim]['ssim']):.4f}")


# =========================
# Visual comparison
# =========================
titles = [
    "Original",
    "Blurred + Noise",
    "TV + FISTA",
    "TV + Split Bregman"
]

plt.figure(figsize=(15, 4))
for i, img in enumerate(visual_results):
    plt.subplot(1, 4, i + 1)
    plt.imshow(img, cmap="gray", vmin=0, vmax=1)
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()


# =========================
# Convergence comparison
# =========================
plt.figure(figsize=(9, 6))

plt.semilogy(results["TV_FISTA"]["history"][0], label="TV + FISTA")
plt.semilogy(results["TV_SplitBregman"]["history"][0], label="TV + Split Bregman")

plt.xlabel("Iteration")
plt.ylabel("Objective Value (log scale)")
plt.title("Convergence Comparison: TV-FISTA vs TV-Split Bregman")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
